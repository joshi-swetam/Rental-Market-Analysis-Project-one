```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#define path
filepath_2018 = "Average_rents_Clean_2018.csv"
filepath_2019 = "Average_rents_Clean_2019.csv"
filepath_2020 = "Average_rents_Clean_2020.csv"
filepath_2021 = "Average_rents_Clean_2021.csv"
filepath_2022 = "Average_rents_Clean_2022.csv"
```

```python
#read files
df_2019 = pd.read_csv(filepath_2019)
df_2020 = pd.read_csv(filepath_2020)
df_2021 = pd.read_csv(filepath_2021)
df_2022 = pd.read_csv(filepath_2022)
df_2018 = pd.read_csv(filepath_2018)
```

```python
#tried merging but will use concat
combined_df = pd.merge(df_2019,df_2020, on= ["Province", "Centre", "Zone", "Neighbourhood"], how = "left")
```

```python
combined_df
```

```python
#combine the data
AR_combined_df = pd.concat([df_2018,df_2019,df_2020,df_2021,df_2022],axis = 0)
AR_combined_df =AR_combined_df.fillna(0)
```

```python
AR_combined_df.to_csv("Average Rents_Concat_.csv",index=False)
```
